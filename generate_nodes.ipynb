```python
from collections import defaultdict
# node_dict_in = defaultdict(list)
node_dict_out = defaultdict(list)

link_indptr = link_data_sparse.indptr.tolist()
indptr_idx = 1
# start = 0
flag = link_indptr[indptr_idx]
flag_pre = -1
for idx, value in enumerate(link_data_sparse.indices.tolist()):
    print(f"idx, value : {idx}, {value}")
    # flag_pre = flag
    
    
    
    while idx == flag:
        indptr_idx+=1
        flag_pre = flag
        flag = link_indptr[indptr_idx]
        
    node_dict_out[indptr_idx-1].append(value)
    
    
```

```python
link_data_sparse_T = sparse.csr_matrix(link_data_reshaped.T)
```

```python

node_dict_in = defaultdict(list)

link_indptr = link_data_sparse_T.indptr.tolist()
indptr_idx = 1
# start = 0
flag = link_indptr[indptr_idx]
flag_pre = -1
for idx, value in enumerate(link_data_sparse_T.indices.tolist()):
    print(f"idx, value : {idx}, {value}")
    # flag_pre = flag
    
    
    
    while idx == flag:
        indptr_idx+=1
        flag_pre = flag
        flag = link_indptr[indptr_idx]
        
    node_dict_in[indptr_idx-1].append(value)
    
    
```
