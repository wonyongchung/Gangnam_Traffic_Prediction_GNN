```python
traffic_220701_tmp = traffic_220701[['시점명', '종점명', '방향', '차선수']]
```

```python
traffic_220701_tmp = traffic_220701_tmp.dropna()
```

```python
road_start_list = traffic_220701_tmp['시점명'].unique()
road_end_list = traffic_220701_tmp['종점명'].unique()
```

```python
road_list = set(np.hstack([road_start_list, road_end_list]))
```

```python
name_idx = dict()
```

```python
for idx, name in enumerate(road_list):
    name_idx[name] = idx
```

```python
from collections import defaultdict
```

```python
link_data = defaultdict(list)
```

```python
for idx in range(len(traffic_220701_tmp)):
    road = traffic_220701_tmp.iloc[idx, :]
    if road['차선수'] == 0:
        road['차선수'] = 1
    if road['방향'] == '하행':
        link_data[name_idx[road['시점명']]].append([name_idx[road['종점명']], np.log(road['차선수']+1)/np.log(8)])
    else:
        link_data[name_idx[road['종점명']]].append([name_idx[road['시점명']], np.log(road['차선수']+1)/np.log(8)])
        
```

```python
from random import sample
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def visualization(sedata, nodedata, samplenum):
    SE = pd.read_csv(f'data/{sedata}', sep=' ', header=0)
    node_idx = pd.read_csv(f'data/{nodedata}', index_col=[0])

    node_randomsample = SE.sample(n = samplenum)

    pca = PCA(n_components=2)
    x = np.array(node_randomsample)
    pcanode = pca.fit_transform(x)
    pcaarray = np.hstack((np.array(node_randomsample.index).reshape(-1,1), pcanode))
    pcadataframe = pd.DataFrame(pcaarray)
    pcadataframe["name"] = ""
    for i in range(len(pcadataframe)):
        pcadataframe.iloc[i, 3] = node_idx.iloc[int(pcadataframe[0][i]), :]['0']
        
    plt.rc('font', family='Malgun Gothic')
            
    plt.figure(figsize=(100,100))
    y = np.array(pcadataframe[2])
    x = np.array(pcadataframe[1])
    n = np.array(pcadataframe["name"])

    fig, ax = plt.subplots()
    ax.scatter(x, y, s= 3)

    for i, txt in enumerate(n):
        ax.annotate(txt, (x[i], y[i]), fontsize = 7)
    plt.savefig(f'data/{sedata[:-4]}_{samplenum}.png')
    plt.show()
```

```python
sedata = "SE160.txt"
nodedata = "node_idx.csv"
samplenum = 30
visualization(sedata, nodedata, samplenum)
```
